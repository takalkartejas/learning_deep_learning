```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import os

```

We are importing CIFAR10 database consisting of variety of objects and we will do an image classification program

```python
dataset = datasets.cifar10.load_data()
```

```python
(X_train, y_train),(X_test, y_test) = dataset
X_train.shape
```

```python
X_test.shape
```

```python
for i in range(5):

    rows, cols =  plt.subplots(1,5)
    cols[i].imshow(X_train[i])

```

```python
y_train[:5]
```

```python
y_train = y_train.reshape(-1,)#lets change it from 3d array to 1d array
y_train[:5]
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
def plot_sample(X,y,index):
    plt.figure(figsize=(15,2))
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]])  #we plot and assign the label using this function
```

```python
plot_sample(X_train, y_train,0)
```

```python
plot_sample(X_train, y_train,1)
```

```python
X_train = X_train/255 #normalize from 0 to 1
X_test = X_test/255
```

```python
ANN = models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(3000, activation = 'relu'),
    layers.Dense(1000, activation = 'relu'),
    layers.Dense(10, activation = 'sigmoid') # we have 10 categories in output layers
])

ANN.compile(optimizer='SGD',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
ANN.fit(X_train, y_train, epochs=5)
```
